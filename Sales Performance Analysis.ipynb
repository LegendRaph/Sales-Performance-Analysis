```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("ratingss.csv")
movies = pd.read_csv("moviess.csv")
tags = pd.read_csv("tagss.csv")
links = pd.read_csv("linkss.csv")
```

```python
d1 = pd.merge(movies, ratings, on="movieId")
d2 = pd.merge(tags, links, on="movieId")
df = pd.merge(d1, d2, on="movieId")
duplicates = df[df.duplicated]
duplicates
```

```python
df.rename(columns={'userId_x' : 'userId_ratings', 'userId_y': 'userId_tags'}, inplace=True)
df.rename(columns={'timestamp_x' : 'ratings_timestamp', 'timestamp_y': 'tags_timestamp'}, inplace=True)
df
```

```python
# Converting to date-time format
def date_time_format(x):
    hours = 0
    minutes = 0
    
    if x//3600 <= 23:
        hours = (x // 3600)
    else:
        hours = (x // 3600) % 24
    if x//60 <= 59:
        minutes = x // 59
    else:
        minutes = (x // 60) % 60
    
    
    if hours <= 9 and minutes <= 9:
        return f"1970-01-01  0{hours}:0{minutes}"
    elif hours <= 9:
        return f"1970-01-01 0{hours}:{minutes}"
    else:
        return f"1970-01-01 {hours}:{minutes}"

df['ratings_timestamp'] = df['ratings_timestamp'].apply(date_time_format)
df['tags_timestamp'] = df['tags_timestamp'].apply(date_time_format)
```

```python
df
```

```python
# New feature 1 - Extract year
def get_year(x):
    split = x.split()
    year = split[-1][1:-1]
    return year

df['year'] = df['title'].apply(get_year)
df
```

```python
# New feature 2 - number of genre
def genre_count(x):
    count = 0
    for i in x:
        if i == '|':
            count = count + 1
    return count + 1


df['Number of Genres'] = df['genres'].apply(genre_count)
```

```python
df
```

```python
# New feature 3 - primary genre
df['primary_genre'] = df['genres'].str.split('|').str[0]
df
```

```python
# New feature 4 - Movie is pg-13?
def pg(x):
    pg13 = False
    gen = x.split('|')
    for i in gen:
        if i == 'Children':
            pg13 = True
    return pg13
    

df['is_pg13'] = df['genres'].apply(pg)
```
